```python
#load packages and data
import pandas as pd
import plotly.graph_objects as go
mpg=pd.read_csv('mpg.csv')
mpg_df=pd.DataFrame(mpg)
print(mpg_df.head())
```

```python
#visualize the correlation with a heatmap
mpg_df_corr=mpg_df.corr(method='pearson')
print(mpg_df_corr)
fig=go.Figure(
    go.Heatmap(
        z=mpg_df_corr.values.tolist(),
        x=mpg_df_corr.columns,
        y=mpg_df_corr.columns,
        colorscale='rdylgn'
    ))

fig.update_layout(
    title='Interactive Correlation Plot',
    legend_title='Legend Title',
    font=dict(
        family='sans_serif,sans_serif',
        size=18,
        color='RebeccaPurple'
    ),
    margin=dict(l=0,r=0,b=0)
)

fig.show(config = {"displayModeBar": False})
```
